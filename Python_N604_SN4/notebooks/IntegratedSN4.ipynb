```python
%matplotlib widget

from orcs.process import SpectralCube
import pylab as pl
import orb.utils.io
import numpy as np
import matplotlib.pyplot as plt
from uncertainties import ufloat

plt.rcParams.update({'font.size': 12, 'xtick.minor.visible': True,
                     'ytick.minor.visible': True, 'xtick.direction': 'in',
                     'ytick.direction': 'in',
                     'pgf.texsystem': 'pdflatex',
                     'font.family': 'serif',
                     'text.usetex': True,
                     'pgf.rcfonts': False})

cube = SpectralCube('/Volumes/RayPass/Research/SIGNALS/N604_SN4/3266414z.hdf5')
```

```python
background = cube.get_spectrum_in_annulus(331,1374,1,40, median=True, mean_flux=True)

region_file = '/Volumes/RayPass/Research/SIGNALS/N604_SN4/ngc604sn4.reg'

axis, sky = cube.extract_integrated_spectrum(region_file, median=True, mean_flux=True)
#pl.plot(axis, sky)
spectrum = cube.get_spectrum_from_region(region_file, median=False, mean_flux=False)
spectrum_corr = spectrum.subtract_sky(background)

lam0 = 6562.819*(1 + ((-214)/2.998e5))

def wave2vel(x):
    return 2.999e5*((x - lam0)/lam0) + (-214)

def vel2wave(x):
    return ((x - (-214))*lam0)/(2.998e5) + lam0
```

```python
0.5*np.nanmax(spectrum_corr.data.real)/1e-12
```

```python
fit1 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548'],
                        fmodel='sincgaussphased',
                        pos_def=['1'],
                        pos_cov=[-214],
                        sigma_def=['1'],
                        sigma_cov=[20],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'])

print(fit1)
fit2 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548','Halpha', '[NII]6583','[NII]6548'],
                        fmodel='sincgaussphased',
                        pos_def=['1','1','1','2','2','2'],
                        pos_cov=[-214,-180],
                        sigma_def=['1','1','1','2','2','2'],
                        sigma_cov=[20,20],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'])

print(fit2)
fit3 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548','Halpha', '[NII]6583','[NII]6548','Halpha',],
                        fmodel='sincgaussphased',
                        pos_def=['1','1','1','2','2','2','3'],
                        pos_cov=[-214,-180,-230],
                        sigma_def=['1','1','1','2','2','2','3'],
                        sigma_cov=[20,20,20],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'])

print(fit3)


fig = plt.figure(figsize=(15,5), constrained_layout=True)
ax = fig.subplot_mosaic(
    """
    abc
    """)

for a,f in zip(['a','b','c'],[fit1,fit2,fit3]):
    ax[a].plot(1e8/axis, spectrum_corr.data/1e-12, color='k', ls='-', lw=1)
    ax[a].set_xlim((6530,6590))
    ax[a].axhline(0, color='k', ls='--', alpha=0.5)
    ax[a].set_xlabel(r'Wavelength [Å]')
    if a == 'a':
        ax[a].set_ylabel(r'Flux [$10^{-12}$ erg s$^{-1}$ cm$^{-2}$ Å$^{-1}$]')

    secax = ax[a].secondary_xaxis('top', functions=(wave2vel,vel2wave))
    secax.set_xlabel(r'Velocity [km/s]')

    y_max = np.nanmax(spectrum_corr.data.real/1e-12) + 0.25*np.nanmax(spectrum_corr.data.real/1e-12)
    y_min = 0 - 0.5*np.nanmax(spectrum_corr.data.real/1e-12)
    ax[a].set_ylim(y_min,y_max)

    ax[a].plot(1e8/axis, f.get_spectrum().data/1e-12, c='red', lw=1, ls='-', label='Fit')
    ax[a].plot(1e8/axis, ((spectrum_corr.data - f.get_spectrum().data)/1e-12) - 5,
               color='k', ls='-', lw=1)

    if a == 'a':
        ax[a].set_title(r'One Component')
        flux = ufloat(f['flux'][0],f['flux_err'][0])
        vel = ufloat(f['velocity'][0],f['velocity_err'][0])
        sig = ufloat(f['broadening'][0],f['broadening_err'][0])
        comp_string = '\n'.join((
            r'Component 1:',
            r'$F = $ {:.2u} erg/s/cm$^{{2}}$',
            r'$V = $ {:.2u} km/s',
            r'$\sigma = $ {:.2u} km/s')).format(flux,vel,sig)
        ax[a].text(0.05,0.95,comp_string, transform=ax[a].transAxes, fontsize=10, va='top')

    if a == 'b':
        i = 0
        comp0 = np.zeros_like(spectrum_corr.data.real)
        comp1 = np.zeros_like(spectrum_corr.data.real)
        for i, imod in zip(range(6), f['fitted_models']['Cm1LinesModel']):
            if i<3: comp0 += imod
            else: comp1 += imod

        ax[a].plot(1e8/axis, comp0/1e-12, c='blue', lw=1, ls='--')
        ax[a].plot(1e8/axis, comp1/1e-12, c='green', lw=1, ls='--')
        ax[a].set_title(r'Two Components')

        flux1 = ufloat(f['flux'][0],f['flux_err'][0])
        vel1 = ufloat(f['velocity'][0],f['velocity_err'][0])
        sig1 = ufloat(f['broadening'][0],f['broadening_err'][0])
        flux2 = ufloat(f['flux'][3],f['flux_err'][3])
        vel2 = ufloat(f['velocity'][3],f['velocity_err'][3])
        sig2 = ufloat(f['broadening'][3],f['broadening_err'][3])
        comp_string = '\n'.join((
            r'Component 1: Blue',
            r'$F = $ {:.2u} erg/s/cm$^{{2}}$',
            r'$V = $ {:.2u} km/s',
            r'$\sigma = $ {:.2u} km/s',
            r'Component 2: Green',
            r'$F = $ {:.2u} erg/s/cm$^{{2}}$',
            r'$V = $ {:.2u} km/s',
            r'$\sigma = $ {:.2u} km/s')).format(flux1,vel1,sig1,flux2,vel2,sig2)
        ax[a].text(0.05,0.95,comp_string, transform=ax[a].transAxes, fontsize=10, va='top')

    
    if a == 'c':
        i = 0
        comp0 = np.zeros_like(spectrum_corr.data.real)
        comp1 = np.zeros_like(spectrum_corr.data.real)
        comp2 = np.zeros_like(spectrum_corr.data.real)
        for i, imod in zip(range(7), f['fitted_models']['Cm1LinesModel']):
            if i<3: 
                comp0 += imod
            elif i in [3,4,5]: 
                # print(i)
                comp1 += imod
            else: 
                comp2 += imod
    
        ax[a].plot(1e8/axis, comp0/1e-12, c='blue', lw=1, ls='--')
        ax[a].plot(1e8/axis, comp1/1e-12, c='green', lw=1, ls='--')
        ax[a].plot(1e8/axis, comp2/1e-12, c='magenta', lw=1, ls='--')
        ax[a].set_title(r'Three Components')

        flux1 = ufloat(f['flux'][0],f['flux_err'][0])
        vel1 = ufloat(f['velocity'][0],f['velocity_err'][0])
        sig1 = ufloat(f['broadening'][0],f['broadening_err'][0])
        flux2 = ufloat(f['flux'][3],f['flux_err'][3])
        vel2 = ufloat(f['velocity'][3],f['velocity_err'][3])
        sig2 = ufloat(f['broadening'][3],f['broadening_err'][3])
        flux3 = ufloat(f['flux'][6],f['flux_err'][6])
        vel3 = ufloat(f['velocity'][6],f['velocity_err'][6])
        sig3 = ufloat(f['broadening'][6],f['broadening_err'][6])
        comp_string = '\n'.join((
            r'Component 1: Blue',
            r'$F = $ {:.2u} erg/s/cm$^{{2}}$',
            r'$V = $ {:.2u} km/s',
            r'$\sigma = $ {:.2u} km/s',
            r'Component 2: Green',
            r'$F = $ {:.2u} erg/s/cm$^{{2}}$',
            r'$V = $ {:.2u} km/s',
            r'$\sigma = $ {:.2u} km/s',
            r'Component 3: Magenta',
            r'$F = $ {:.2u} erg/s/cm$^{{2}}$',
            r'$V = $ {:.2u} km/s',
            r'$\sigma = $ {:.2u} km/s')).format(flux1,vel1,sig1,flux2,vel2,sig2,flux3,vel3,sig3)
        ax[a].text(0.05,0.95,comp_string, transform=ax[a].transAxes, fontsize=10, va='top')

plt.savefig('Plots/IntegratedSpectrum.png', bbox_inches='tight')
    
print('=== BIC COMPARISON ===')
print('if positive and > 2, two component fit is better')
print("""0 to 2 \t\tNot worth more than a bare mention
2 to 6 \t\tPositive
6 to 10 \tStrong
>10 \t\tVery strong """)

print('DELTA BIC 1 to 2=', fit1['BIC'] - fit2['BIC'])
print('DELTA BIC 2 to 3=', fit2['BIC'] - fit3['BIC'])
```

```python
from cmcrameri import cm
from astropy.visualization import ImageNormalize, ManualInterval, LogStretch
from astropy.wcs import WCS
from astropy.io import fits
from regions import Regions
from astropy.coordinates import SkyCoord
import astropy.units as u

hdu = fits.open('../../Python_N604/NGC604_SN4/MAPS_OneComponent/NGC604_SN4.LineMaps.map.6563.1x1.flux.fits')
halpha = hdu[0].data
hdr = hdu[0].header
wcs = WCS(hdr)

gal_center = SkyCoord(23.637958, 30.785204, unit='deg',frame='icrs')

reg = Regions.read(region_file, format='ds9')[0]
regions_sky = reg.to_sky(wcs)
regions_pix = regions_sky.to_pixel(wcs)

fig = plt.figure(figsize=(7,6), constrained_layout=True)
ax = fig.subplot_mosaic(
    """
    a
    """,
    subplot_kw={'projection': wcs})

norm = ImageNormalize(halpha/1e-15, interval=ManualInterval(vmin=1e-17/1e-15, vmax=1e-14/1e-15), stretch=LogStretch())
t = ax['a'].imshow(halpha/1e-15, origin='lower', norm=norm, cmap=cm.lajolla)
cbar = plt.colorbar(t, ax=ax['a'])
cbar.set_label(r'H$\alpha$ Flux [$10^{-15}$ erg s$^{-1}$ cm$^{-2}$ arcsec$^{-1}$]')
regions_pix.plot(ax=ax['a'], fc='None', ec='limegreen', lw=2)
# ax['a'].contour(halpha.T, levels=np.logspace(-16.7,-14,6), colors='k', linestyles='solid', linewidths=1)

ax['a'].set_xlim(825,1280)
ax['a'].set_ylim(730,1310)

ra = ax['a'].coords["ra"]
dec = ax['a'].coords["dec"]
ra.set_ticks_visible(False)
ra.set_ticklabel_visible(False)
dec.set_ticks_visible(False)
dec.set_ticklabel_visible(False)
ra.set_axislabel("")
dec.set_axislabel("")
aframe = gal_center.skyoffset_frame()
overlay = ax['a'].get_coords_overlay(aframe)
ra_offset = overlay["lon"]
dec_offset = overlay["lat"]
dec_offset.set_ticks(spacing=20 * u.arcsec, color='white')
ra_offset.set_ticks(spacing=20 * u.arcsec, color='white')
overlay.grid()
ra_offset.set_major_formatter("s.s")
dec_offset.set_major_formatter("s.s")
ra_offset.set_ticks_position("bt")
ra_offset.set_ticklabel_position("b")
dec_offset.set_ticks_position("lr")
dec_offset.set_ticklabel_position("l")
ra_offset.set_axislabel_position("b")
dec_offset.set_axislabel_position("l")
dec_offset.set_axislabel(r"$\Delta\delta$")
ra_offset.set_axislabel(r"$\Delta\alpha$")

ax['a'].tick_params(direction='in', color='white')
# ax['a'].grid()
plt.savefig('Plots/IntegratedRegion.png', bbox_inches='tight')
```

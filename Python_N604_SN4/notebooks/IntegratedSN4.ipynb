```python
%matplotlib widget

from orcs.process import SpectralCube
import pylab as pl
import orb.utils.io
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 12, 'xtick.minor.visible': True,
                     'ytick.minor.visible': True, 'xtick.direction': 'in',
                     'ytick.direction': 'in',
                     'pgf.texsystem': 'pdflatex',
                     'font.family': 'serif',
                     'text.usetex': True,
                     'pgf.rcfonts': False})

cube = SpectralCube('/Volumes/RayPass/Research/SIGNALS/N604_SN4/3266414z.hdf5')
```

```python
background = cube.get_spectrum_in_annulus(331,1374,1,40, median=True, mean_flux=True)

region_file = '/Volumes/RayPass/Research/SIGNALS/N604_SN4/ngc604sn4.reg'

axis, sky = cube.extract_integrated_spectrum(region_file, median=True, mean_flux=True)
#pl.plot(axis, sky)
spectrum = cube.get_spectrum_from_region(region_file, median=False, mean_flux=False)
spectrum_corr = spectrum.subtract_sky(background)

fit1 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548'],
                        fmodel='sincgaussphased',
                        pos_def=['1'],
                        pos_cov=[-214],
                        sigma_def=['1'],
                        sigma_cov=[20],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'])

print(fit1)
fit2 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548','Halpha', '[NII]6583','[NII]6548'],
                        fmodel='sincgaussphased',
                        pos_def=['1','1','1','2','2','2'],
                        pos_cov=[-214,-180],
                        sigma_def=['1','1','1','2','2','2'],
                        sigma_cov=[20,20],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'])

print(fit2)
fit3 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548','Halpha', '[NII]6583','[NII]6548','Halpha',],
                        fmodel='sincgaussphased',
                        pos_def=['1','1','1','2','2','2','3'],
                        pos_cov=[-214,-180,-230],
                        sigma_def=['1','1','1','2','2','2','3'],
                        sigma_cov=[20,20,20],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'])

print(fit3)


fig = plt.figure(figsize=(15,5), constrained_layout=True)
ax = fig.subplot_mosaic(
    """
    abc
    """)

for a,f in zip(['a','b','c'],[fit1,fit2,fit3]):
    ax[a].plot(1e8/axis, spectrum_corr.data/1e-11, color='k', ls='-', lw=1)
    ax[a].set_xlim((6530,6590))
    ax[a].axhline(0, color='k', ls='--', alpha=0.5)
    ax[a].set_xlabel(r'Wavelength [Å]')
    if a == 'a':
        ax[a].set_ylabel(r'Flux [$10^{-11}$ erg s$^{-1}$ cm$^{-2}$ Å$^{-1}$]')

    ax[a].plot(1e8/axis, f.get_spectrum().data/1e-11, c='red', lw=1, ls='-', label='Fit')
    ax[a].plot(1e8/axis, ((spectrum_corr.data - f.get_spectrum().data)/1e-11) - 0.5,
               color='k', ls='-', lw=1)

    if a == 'b':
        i = 0
        comp0 = np.zeros_like(spectrum_corr.data.real)
        comp1 = np.zeros_like(spectrum_corr.data.real)
        for i, imod in zip(range(6), fit2['fitted_models']['Cm1LinesModel']):
            if i<3: comp0 += imod
            else: comp1 += imod

        ax[a].plot(1e8/axis, comp0/1e-11, c='blue', lw=1, ls='--')
        ax[a].plot(1e8/axis, comp1/1e-11, c='green', lw=1, ls='--')

    
print('=== BIC COMPARISON ===')
print('if positive and > 2, two component fit is better')
print("""0 to 2 \t\tNot worth more than a bare mention
2 to 6 \t\tPositive
6 to 10 \tStrong
>10 \t\tVery strong """)

print('DELTA BIC 1 to 2=', fit1['BIC'] - fit2['BIC'])
print('DELTA BIC 2 to 3=', fit2['BIC'] - fit3['BIC'])
```

```python
import gvar

velocity1_gvar = gvar.gvar(-214, 5)
velocity2_gvar = gvar.gvar(-235, 10)
velocity3_gvar = gvar.gvar(20, 5)
broadening1_gvar = gvar.gvar(20, 5)
broadening2_gvar = gvar.gvar(20, 10)
broadening3_gvar = gvar.gvar(50, 10)

fit1 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548'],
                        fmodel='sincgaussphased',
                        pos_def=['1'],
                        pos_cov=[-214],
                        sigma_def=['1'],
                        sigma_cov=[50],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'])

print(fit1)

fit2 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548','Halpha','[NII]6583','[NII]6548'],
                        fmodel='sincgaussphased',
                        pos_def=['1','1','1','2','2','2'],
                        pos_cov=[velocity1_gvar, velocity2_gvar],
                        sigma_def=['1','1','1','2','2','2'],
                        sigma_guess=[broadening1_gvar, broadening1_gvar, broadening1_gvar, broadening2_gvar, broadening2_gvar, broadening2_gvar],
                        alpha_cov=[np.pi/20],
                        alpha_def=['1'],
                        snr_guess=[85,22,6,85,22,6])

print(fit2)

fit3 = spectrum_corr.fit(['Halpha', '[NII]6583','[NII]6548','Halpha','[NII]6583','[NII]6548','Halpha'],
                    fmodel='sincgaussphased',
                    pos_def=['1','1','1','2','2','2','3'],
                    pos_cov=[velocity1_gvar, velocity2_gvar, velocity3_gvar],
                    sigma_def=['1','1','1','2','2','2','3'],
                    sigma_guess=[broadening1_gvar, broadening1_gvar, broadening1_gvar, broadening2_gvar, broadening2_gvar, broadening2_gvar, broadening3_gvar],
                    alpha_cov=[np.pi/20],
                    alpha_def=['1'],
                    snr_guess=[85,22,6,85,22,6])

print(fit3)
```

```python
lam0 = 6562.819*(1 + (fit1['velocity'][0]/2.998e5))

fig = plt.figure(figsize=(15,5), constrained_layout=True)
ax = fig.subplot_mosaic(
    """
    abc
    """)

for a,f in zip(['a','b','c'],[fit1,fit2,fit3]):
    ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), spectrum_corr.data/1e-11, color='k', ls='-', lw=1)
    ax[a].set_xlim((-500,500))
    ax[a].axhline(0, color='k', ls='--', alpha=0.5)
    ax[a].set_xlabel(r'Velocity Offset [km s$^{-1}$]')
    if a == 'a':
        ax[a].set_ylabel(r'Flux [$10^{-11}$ erg s$^{-1}$ cm$^{-2}$ Å$^{-1}$]')

    ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), f.get_spectrum().data/1e-11, c='red', lw=1, ls='-', label='Fit')
    ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), ((spectrum_corr.data - f.get_spectrum().data)/1e-11) - 0.5,
               color='k', ls='-', lw=1)

    if a == 'b':
        models = f['fitted_models']['Cm1LinesModel']
        ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), models[0]/1e-11, c='blue', lw=1, ls='--')
        ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), models[3]/1e-11, c='green', lw=1, ls='--')

    if a == 'c':
        models = f['fitted_models']['Cm1LinesModel']
        ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), models[0]/1e-11, c='blue', lw=1, ls='--')
        ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), models[3]/1e-11, c='green', lw=1, ls='--')
        ax[a].plot(2.998e5*(((1e8/axis) - lam0)/lam0), models[6]/1e-11, c='magenta', lw=1, ls='--')

print('=== BIC COMPARISON ===')
print('if positive and > 2, two component fit is better')
print("""0 to 2 \t\tNot worth more than a bare mention
2 to 6 \t\tPositive
6 to 10 \tStrong
>10 \t\tVery strong """)

print('DELTA BIC 1 to 2=', fit1['BIC'] - fit2['BIC'])
print('DELTA BIC 2 to 3=', fit2['BIC'] - fit3['BIC'])
```
